- Calculate the phi torsions of a trajectory:

```python
phi_atoms, phi_traj = md.compute_phi(trajectory)
```

```python
phi_traj.shape
```

- Calculate the psi torsions of a trajectory:

```python
psi_atoms, psi_traj = md.compute_psi(trajectory)
```

```python
psi_traj.shape
```

```python
# phi

plt.plot(trajectory.time, phi_traj[:,0],'g')
plt.xlabel('time')
plt.ylabel('phi torsions')
plt.show()
```

```python
# psi

plt.plot(trajectory.time, psi_traj[:,0], 'g')
plt.xlabel('time')
plt.ylabel('psi torsions')
plt.show()
```

```python
# phi vs psi

plt.scatter(phi_traj[:,0], psi_traj[:,0])
plt.xlabel('phi torsions')
plt.ylabel('psi torsions')
plt.ylim(-np.pi, np.pi)
plt.xlim(-np.pi, np.pi)
plt.show()
```

```python
histo_matrix, xedges, yedges, histo = plt.hist2d(phi_traj[:,0], psi_traj[:,0], bins=40, range = [[-np.pi, np.pi], [-np.pi, np.pi]],
                                                cmap = "Greys")
```
